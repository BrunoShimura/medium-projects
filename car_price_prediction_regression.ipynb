```python
# Defining a mapping dictionary
fueltype_mapping = {'gas': 1, 'diesel': 2}
enginelocation_mapping = {'front': 1, 'rear': 2}
fuelsystem_mapping = {'mpfi': 1,'2bbl': 2, 'mfi': 3, '1bbl': 4, 'spfi': 5, '4bbl': 6, 'idi': 7, 'spdi': 8}
aspirationtype_mapping = {'std': 1, 'turbo': 2}
carbody_mapping = {'convertible': 1, 'hatchback': 2, 'sedan': 3, 'wagon': 4, 'hardtop': 5}
drivewheel_mapping = {'rwd': 1, 'fwd': 2, '4wd': 3}
doornumber_mapping = {'two': 2, 'four': 4}
enginetype_mapping = {'dohc': 1, 'ohcv': 2, 'ohc': 3, 'l': 4, 'rotor': 5, 'ohcf': 6, 'dohcv': 7}
cylindernumber_mapping = {'four': 4, 'six': 6, 'five': 6, 'three': 3, 'twelve': 12, 'two': 2, 'eight': 8}

#Using the pat function to apply the mapping to all columns
df['fueltype'] = df['fueltype'].map(fueltype_mapping)
df['enginelocation'] = df['enginelocation'].map(enginelocation_mapping)
df['fuelsystem'] = df['fuelsystem'].map(fuelsystem_mapping)
df['aspiration'] = df['aspiration'].map(aspirationtype_mapping)
df['carbody'] = df['carbody'].map(carbody_mapping)
df['drivewheel'] = df['drivewheel'].map(drivewheel_mapping)
df['doornumber'] = df['doornumber'].map(doornumber_mapping)
df['enginetype'] = df['enginetype'].map(enginetype_mapping)
df['cylindernumber'] = df['cylindernumber'].map(cylindernumber_mapping)
```

```python
df.info()
```

```python
df["enginelocation"].unique()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import numpy as np
```

```python
X = df.drop(['CarName', 'price', 'symboling', 'fueltype', 'aspiration', 'carbody', 'enginelocation', 'enginetype', 'fuelsystem', 'stroke', 'compressionratio', 'peakrpm', 'carheight','doornumber'], axis='columns')
y = df['price']
```

```python
print(X.shape, y.shape)
```

```python
# train 60%
X_train, X_, y_train, y_ = train_test_split(X, y, test_size=0.40, random_state=1)
```

```python
# test 20% and validation 20%
X_cv, X_test, y_cv, y_test = train_test_split(X_, y_, test_size=0.50, random_state=1)
```

```python
# Initialize the class
linear_model = LinearRegression()
# Train the model
linear_model.fit(X_train, y_train )
```

```python
# Feed the scaled training set and get the predictions
yhat = linear_model.predict(X_train)
print(f"training MSE: {mean_absolute_error(y_train, yhat)}")
```

```python
# Feed the scaled cross validation set
yhatcv = linear_model.predict(X_cv)
# Use scikit-learn's utility function and divide by 2
print(f"Cross validation MSE: {mean_absolute_error(y_cv, yhatcv)}")
```

```python
linear_model.score(X_train, y_train)
```

```python
predict = linear_model.predict(X_test)
```

```python
mean_absolute_error(predict, y_test)
```

```python
import numpy as np
np.subtract(predict, y_test).mean()
```
